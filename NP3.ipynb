```python
import numpy as np
# 1. Configuration
num_students = 500
filename = 'student_records.csv'

# 2. Generate Unique 8-digit Roll Numbers (from 10,000,000 to 99,999,999)
roll_numbers = np.random.randint(10000000, 100000000, size=(num_students, 1))

# 3. Generate Marks (0-100) for 5 subjects: 
marks = np.random.randint(0, 101, size=(num_students, 5))

# 4. Combine into one matrix
data = np.hstack([roll_numbers, marks])

# 5. Define headers
header = "Roll_No,Physics,Mathematics 1,Drawing,Electrical Engineering,Programming"

# 6. Save to CSV
np.savetxt(filename, data, delimiter=',', header=header, comments='', fmt='%d')

print(f"Created {filename} with 500 records and updated subject names.")
```
