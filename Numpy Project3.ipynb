```python
import numpy as np
students=np.genfromtxt('student_records.csv',delimiter=',',skip_header=True)
#to find the total marks of each student, we can sum the marks of all subjects for each student
total_marks=students[:,1:6].sum(axis=1)
students=np.column_stack((students,total_marks))
#to find the percentage of each student, we can divide the total marks by the maximum marks (500) and multiply by 100
percentages=(total_marks/500)*100
percentages=np.round(percentages,2)
students=np.column_stack((students,percentages))
print(students)
```

```python
#to find the average marks of each subject, we can calculate the mean of each column for the marks of all students
average_marks=students[:,1:6].mean(axis=0)
print("Average marks for each subject:", average_marks)
```

```python
#the code that helps us find the good and bad performers in each subject and overall
physics_good=students[students[:,1]>85]
physics_bad=students[students[:,1]<35]
maths_good=students[students[:,2]>85]
maths_bad=students[students[:,2]<35]
drawing_good=students[students[:,3]>85]
drawing_bad=students[students[:,3]<35]
electrical_good=students[students[:,4]>85]
electrical_bad=students[students[:,4]<35]
programming_good=students[students[:,5]>85]
programming_bad=students[students[:,5]<35]
overall_good=students[students[:,-1]>75]
overall_bad=students[students[:,-1]<35]
```

```python
#the code that prints the good and bad performers in each subject and overall
print("Good performers in Physics:\n", physics_good)
print("Bad performers in Physics:\n", physics_bad)
print("Good performers in Maths:\n", maths_good)
print("Bad performers in Maths:\n", maths_bad)
print("Good performers in Drawing:\n", drawing_good)
print("Bad performers in Drawing:\n", drawing_bad)
print("Good performers in Electrical:\n", electrical_good)
print("Bad performers in Electrical:\n", electrical_bad)
print("Good performers in Programming:\n", programming_good)
print("Bad performers in Programming:\n", programming_bad)
print("Overall good performers:\n", overall_good)
print("Overall bad performers:\n", overall_bad)
```

```python
#code to add the records of students to a new file along with their total marks and percentages
np.savetxt('student_records_with_totals.csv', students, delimiter=',', fmt='%s', header='ID,Physics,Maths,Drawing,Electrical Engineering,Programming,Total Marks,Percentage', comments='')
print("file saved successfully with total marks and percentages.")
```
